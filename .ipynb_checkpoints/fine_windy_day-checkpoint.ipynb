### FEATURE ENGINEERING

### Missing Values

```python
### Handing categorical features with missing values
```

```python
dataset[categorical].isnull().sum()
```

```python
features_cat_nan = [feature for feature in categorical if dataset[feature].isnull().sum()>1]
features_cat_nan
```

```python
def replace_cat_features(dataset,features_with_nan):
    data = dataset.copy()
    data[features_with_nan] = data[features_with_nan].fillna('Missing')
    return data
dataset = replace_cat_features(dataset,features_cat_nan)
dataset[categorical].head()
```

```python
dataset[categorical].isnull().sum()
```

```python
dataset.head()
```

```python
### Handling numerical features with missing values
```

```python
features_num_nan = [feature for feature in numerical_features if dataset[feature].isnull().sum()>1]
len(features_num_nan)
```

```python
for feature in features_num_nan:
    median = dataset[feature].median()
    
    dataset[feature+'nan'] = np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median,inplace=True)

dataset[features_num_nan].isnull().sum()
```

```python
dataset.head()
```

```python
dataset['cloud_level'].unique()
```

```python
dataset['cloud_level'] = dataset['cloud_level'].replace(['Missing','Extremely Low','Low','Medium'],[0,1,2,3])
```

```python
dataset.head()
```

```python
turbine_status = dataset['turbine_status']
turbine_status = pd.get_dummies(turbine_status,drop_first = True)
turbine_status.head()
```

```python
dataset.shape
```

```python
turbine_status.shape
```

```python
dataset = pd.concat([dataset,turbine_status],axis=1)
```

```python
dataset.head()
```

```python
dataset.drop(['turbine_status'],axis=1,inplace=True)
```

```python
dataset.dtypes
```

```python
feature_scale = [feature for feature in dataset.columns if feature not in ['tracking_id','datetime','windmill_generated_power(kW/h)']]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset[feature_scale])
```

```python
scaler.transform(dataset[feature_scale])
```

```python
dataset = dataset.dropna()
```

```python
x_train = pd.DataFrame(scaler.transform(dataset[feature_scale]), columns = feature_scale)
x_train.head()
```

```python
x_train.isnull().sum()
```

```python
x_train.shape
```

```python
x_train = x_train.dropna()
```

### Feature Selection

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(x_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
selected_features = x_train.columns[(feature_sel_model.get_support())]
print("Total NUmber of features : ",x_train.shape[1])
print("Number of selected features : ",len(selected_features))
print("Features with coefficients shrunk to 0 : ",np.sum(feature_sel_model.estimator_.coef_ == 0))
```

```python
selected_features
```

```python
x_train = x_train[selected_features]
```
